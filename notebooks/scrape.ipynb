```python
### Property with 1 Bedroom

#pip install requests
#!pip install html5lib
#pip install bs4
from json import dump
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
import re

#Constants
BASE_URL = "https://www.domain.com.au"
BED1_URL = "/rent/melbourne-region-vic/?bedrooms=1&sort=dateupdated-desc&page="
N_PAGES = range(1, 50) # update this to your liking

#Access HTML content from webpage
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}

url_links = []
property_metadata = defaultdict(dict)

for page in N_PAGES:
    url = BASE_URL + BED1_URL + f"{page}"
    r = requests.get(url = url, headers=headers)
    bs_object = BeautifulSoup(r.content, "html.parser")

    # find the unordered list (ul) elements which are the results, then
    # find all href (a) tags that are from the base_url website.
    index_links = bs_object \
        .find(
            "ul",
            {"data-testid": "results"}
        ) \
        .findAll(
            "a",
            href=re.compile(f"{BASE_URL}/*") # the `*` denotes wildcard any
        )
    for link in index_links:
        # if its a property address, add it to the list
        if 'address' in link['class']:
            url_links.append(link['href'])

# for each url, scrape some basic metadata
for property_url in url_links[1:]:
    url = property_url
    r = requests.get(url = url, headers=headers)
    bs_object = BeautifulSoup(r.content, "html.parser")

    # looks for the header class to get property name
    try: property_metadata[property_url]['name'] = bs_object \
        .find("h1", {"class": "css-164r41r"}) \
        .text
    except: property_metadata[property_url]['name'] = None

    # looks for the div containing a summary title for cost
    try: property_metadata[property_url]['cost_text'] = bs_object \
        .find("div", {"data-testid": "listing-details__summary-title"}) \
        .text      
    except: property_metadata[property_url]['cost_text'] = None

    # extract coordinates from the hyperlink provided
    try: property_metadata[property_url]['coordinates'] = [
        float(coord) for coord in re.findall(
            r'destination=([-\s,\d\.]+)', # use regex101.com here if you need to
            bs_object \
                .find(
                    "a",
                    {"target": "_blank", 'rel': "noopener noreferer"}
                ) \
                .attrs['href']
        )[0].split(',')
    ]
    except: property_metadata[property_url]['coordinates'] = None

    #Property room and size
    try: property_metadata[property_url]['rooms'] = [
        feature.text for feature in bs_object \
            .find("div", {"data-testid": "property-features"}) \
            .findAll("span", {"data-testid": "property-features-text-container"})
    ]
    except: property_metadata[property_url]['rooms'] = None
    #Property description
    try: property_metadata[property_url]['desc'] = re \
        .sub(r'<br\/>', '\n', str(bs_object.find("p"))) \
        .strip('</p>') 
    except: property_metadata[property_url]['desc'] = None      
    #Available date
    try: property_metadata[property_url]['available_date'] = bs_object \
        .find("div", {"class":"css-hboir5"}) \
        .find("strong") \
        .text
    except:
        property_metadata[property_url]['available_date'] = None

    #Property type
    try: property_metadata[property_url]['property_type'] = [
        feature.text for feature in bs_object \
            .find("div", {"data-testid": "listing-summary-property-type"}) \
            .findAll("span")
    ][0]
    except: property_metadata[property_url]['property_type'] = None

    try: property_metadata[property_url]['nearest_schools'] = [
        feature.text for feature in bs_object \
            .find("div", {"data-testid": "fe-co-school-catchment-content"}) \
            .findAll("div", {"data-testid": "fe-co-school-catchment-schoolDistance"})
    ]
    except: property_metadata[property_url]['nearest_schools'] = None

#Save dictionary as CSV
import csv
csv_columns = ['name','cost_text','coordinates','rooms','desc', 'available_date', 'property_type','nearest_schools']
csv_file = "../data/raw/bed1.1.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        #for data in property_metadata:
        for property_url in url_links[1:]:
            writer.writerow(property_metadata[property_url])
except IOError:
    print("I/O error")
```
