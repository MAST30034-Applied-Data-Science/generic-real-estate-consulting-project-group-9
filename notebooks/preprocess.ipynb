# Preprocess

## Basic proprocess

## bed1.1.csv

```python
import pandas as pd
property_df2 = pd.read_csv('../data/raw/bed1.1.csv')
float_coords = []

# change the format of cooredinates
for prop_coord in property_df2["coordinates"]:
    prop_coord2 = prop_coord.split(', ')
    x = float(str(prop_coord2[0][1:]))
    y = float(str(prop_coord2[1][:-1]))
    prop_coord = [x,y]
    float_coords.append(prop_coord)
property_df2["coordinates1"] = float_coords

import re

# derive the postcodes of properties in the scraped names of properties
post_list = []
for name_list in property_df2['name']:
    post_list.append(re.findall(r"3\d\d\d|8\d\d\d|9\d\d\d", name_list))

property_df2['postcode'] = post_list

for i in range(0, len(property_df2)):
    property_df2['postcode'][i] = property_df2['postcode'][i][len(property_df2['postcode'][i]) - 1]

# derive the numbers of room of properties
for i in range(0, len(property_df2)):
    property_df2['rooms'][i] = re.sub(r"\−", '0', property_df2['rooms'][i])

room_num = []
for room_list in property_df2['rooms']:
    room_num.append(re.findall(r"\d+", room_list)[0:3])

property_df2['rooms'] = room_num

drop_list = []
for i in range(0, len(property_df2)):
    if len(property_df2['rooms'][i]) == 0:
        drop_list.append(i)
property_df2 = property_df2.drop(property_df2.index[drop_list]) # remove properties which do not provide information about the numbers of rooms

bed_list = []
for i in range(0, len(property_df2) + len(drop_list)):
    if i not in drop_list:
        bed_list.append(property_df2['rooms'][i][0])
property_df2['bedrooms'] = bed_list

bath_list = []
for i in range(0, len(property_df2) + len(drop_list)):
    if i not in drop_list:
        bath_list.append(property_df2['rooms'][i][1])
property_df2['bathrooms'] = bath_list

park_list = []
for i in range(0, len(property_df2) + len(drop_list)):
    if i not in drop_list:
        park_list.append(property_df2['rooms'][i][2])
property_df2['parking'] = park_list

# derive the distance to the closest school of each property
school_list = []
for schools in property_df2['nearest_schools']:
    school_list.append(re.findall(r"\d\.\d|\d", schools))

property_df2['schools'] = school_list

for i in range(0, len(property_df2) + len(drop_list)):
    if i not in drop_list:
        property_df2['schools'][i] = property_df2['schools'][i][0]

property_df2_2=property_df2[['name', 'postcode', 'property_type', 'bedrooms', 'bathrooms', 'parking','schools', 'cost_text']]

# merge the distance to the closest train station of each property with the dataset
fileObject2 = open("../data/curated/bed1.txt", "r") # read in the text file containing distances and convert it to a list
data2 = fileObject2.read()

data2 = re.sub(r'\n', '',data2)
data2 = re.sub(r'\[', '',data2)
data2 = re.sub(r'\]', '',data2)
data2 = data2.split(', ')

if len(drop_list) != 0:
    for i in drop_list:
        data2.pop(i)


property_df2_2['train_station(m)']=data2

# merge the distance to CBD of each property with the dataset
fileObject3 = open("../data/curated/bed1_distance_cbd.txt", "r") # read in the text file containing distances and convert it to a list
data3 = fileObject3.read()

data3 = re.sub(r'\n', '',data3)
data3 = re.sub(r'\[', '',data3)
data3 = re.sub(r'\]', '',data3)
data3 = data3.split(', ')

if len(drop_list) != 0:
    for i in drop_list:
        data3.pop(i)


property_df2_2['CBD(m)']=data3

property_df2_2.to_csv('../data/curated/bed1.csv') # ouput the dataset after basic preprocess
```

```python
import pandas as pd
df11 = pd.read_csv('../data/curated/final.csv')
```

```python
import re 
outliers2 = []
for i in range(0, len(df11)):
    df11['cost_text'][i] = re.sub(r",",'', df11['cost_text'][i])
    if len(re.findall(r"\$\d+|\d\d+", df11['cost_text'][i])) == 0:  # remove properties which do not provide information of price
        outliers2.append(i)
    if len(re.findall(r"\$\d+|\d\d+", df11['cost_text'][i])) > 0:
        df11['cost_text'][i] = re.findall(r"\$\d+|\d\d+", df11['cost_text'][i])[0]
        df11['cost_text'][i] = re.findall(r"\d+", df11['cost_text'][i])[0]
    
```

```python
# number of properties without price information
len(outliers2)
```

```python
if len(outliers2) > 0:
    df11 = df11.drop(df11.index[outliers2])
```

```python
# output the final version of dataset after prepocess 
df11.to_csv('../data/curated/final2.csv', index=False)
```
