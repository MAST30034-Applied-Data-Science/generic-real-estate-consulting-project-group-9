```python
import pandas as pd
df = pd.read_csv('../data/raw/final2.csv')
df.head()
```

```python
df_1 = df.groupby('postcode').mean()
df_1
```

```python
# top  5 suburb has shortest distance to school
#liveable for family that has kids
df_school = df_1.sort_values(by='schools', axis=0, ascending=True)
df_school.head()
```

```python
# top  5 suburb has shortest distance to CBD
#liveable for youths
df_CBD = df_1.sort_values(by='CBD(m)', axis=0, ascending=True)
df_CBD.head()
```

```python
# top  5 suburb has shortest distance to train station
#liveable for old people who have difficulty with their legs
df_train = df_1.sort_values(by='train_station(m)', axis=0, ascending=True)
df_train.head()
```

```python
#top  5 suburb has most house properties
#more peoperties means more liveable in general case, if do not have any special demand
#limitation:Only the property data extracted from real estate, 
# the extracted property data is not complete， so the number of properties differs from the real data
df_property_num = df.groupby('postcode').count()
df_property_num.sort_values(by = 'name', axis=0, ascending=False).head()
```

```python
# top  5 suburb has highest population
#liveable generally
df_population = df_1.sort_values(by='population2022', axis=0, ascending=False)
df_population.head()
```
